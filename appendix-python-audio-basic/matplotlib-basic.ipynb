# Matplotlib Basics

```python
# import os
import numpy as np
from matplotlib import pyplot as plt
import librosa
import librosa.display
%matplotlib inline
```

### Plot two functions in the same figure

```python
plt.figure(figsize=(10, 3), facecolor='skyblue') # Layout
plt.title('CosSin Figure', loc='center', fontsize=25) # Layout 
X = np.linspace(-np.pi, np.pi, 100, endpoint=True)
# X
# plt.plot(X)
C = np.cos(X)
S = np.sin(X)
plt.plot(X, C, color='blue', linewidth=2.0, linestyle='--', label='cos()')
plt.plot(X, S, color='orange', linewidth=2.0, linestyle='-', label='sin()')
plt.legend(loc='upper right', fontsize=20) # Layout
plt.xlim(-4, 4) # Layout
plt.ylim(-1.5, 1.5) # Layout
plt.xticks([-np.pi, 0, np.pi], [r'$-\pi$', r'$0$', r'$+\pi$'], fontsize=20) # Layout
plt.yticks([-0.8, 0.3, 0.8]) # Layout
plt.xlabel('radians', fontsize=16) # Layout
plt.ylabel('amplitude', fontsize=20) # Layout
plt.tight_layout()
```
