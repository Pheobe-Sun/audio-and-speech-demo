# Signal Visualization

### Signal

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Fs = 256
t_start = 0
t_end = 2
t = np.arange(t_start, t_end, 1/Fs)
x = np.sin(2 * np.pi * 3 * t) +  0.3 * np.sin(2 * np.pi * 17 * t)
```

#### Signal

```python
plt.plot(t, x)
plt.title('Signal $x$', fontsize=20)
plt.xlabel('Time (sec)')
plt.ylabel('Amplitude')
```

#### FFT

```python
X = np.fft.fft(x)
freq = np.fft.fftfreq(x.size, d=1/Fs)
# freq
plt.plot(freq, np.abs(X))
plt.title('Magnitidue DFT $|X|$')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
```

#### DFT

```python
tau = 100
X_mod = X.copy()
X_mod[np.abs(X) < tau] = 0
plt.plot(freq, np.abs(X_mod))
plt.title('Magnitude DFT $|X_\mathrm{mod}|$')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
```

#### ??? Real

```python
x_mod = np.real(np.fft.ifft(X_mod))
# x_mod
plt.plot(t, x_mod)
plt.title('Signal $x_\mathrm{mod}$')
plt.xlabel('Time (sec)')
plt.ylabel('Amplitude')
```
