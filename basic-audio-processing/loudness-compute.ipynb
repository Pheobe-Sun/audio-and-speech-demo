# Compute Loudness

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio
%matplotlib inline
```

```python
Fs = 22050
x, Fs = librosa.load('../audio/narration_44k_24bit_stereo.wav', sr=Fs, mono=True)
win_len_sec = 0.1 # window length in seconds
power_ref = 10**(-12) # reference power level (0 dB)
```

#### Show sound wave and loudness

```python
plt.figure(figsize=(12,2))
plt.plot(x)
plt.tight_layout()
plt.show()
Audio(x, rate=Fs)
```

```python
win_len = round(win_len_sec * Fs)
win = np.ones(win_len) / win_len
power_db = 10 * np.log10(np.convolve(x**2, win, mode='same') / power_ref)
```

```python
plt.figure(figsize=(12,4))
plt.plot(power_db)
plt.ylim([50, 110])
plt.xlabel('Time (seconds)', fontsize=14)
plt.ylabel('Power (dB)', fontsize=14)
plt.tight_layout()
plt.show()

```
